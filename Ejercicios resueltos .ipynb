Ejercicio 1. Conocer si el año que ingresamos fue o será un año bisiesto o no.

```python
año= input('Ingrese el año que desea comprobar')
num_año= int(año)

if num_año %4==0:
  print(num_año,'es un año bisiesto')
else:
  print(num_año, 'no es año bisiesto')
```

Ejercicio 3. Crea una funcion para contar vocales 

```python
def count_vocales (frase):
  count=0
  vocales=['a','e','i','o','u','A','E','I','O','U']
  for char in frase:
    if char in vocales:
      count += 1
  return count
print(count_vocales('Hola Mundo'))
```

Ejercicio 4. Crea una función que elimine los numeros repetidos de las listas de números.

```python
def remove_duplicate(items):
  unique=[]
  for item in items:
    if item not in unique:
      unique.append(item)
  return unique

numbers=[22,11,3,1,4,5,5,2,2,11,66,89]
print(remove_duplicate(numbers))
```

Ejercicio 5. Series de Fibonacci.

```python
def fibonacci(num):
  fibo=[0,1]
  i=2
  while i<= num:
    sig_fibo = fibo[i-1] + fibo[i-2]
    fibo.append(sig_fibo)
    i += 1
  return fibo
print(fibonacci(9))
```
